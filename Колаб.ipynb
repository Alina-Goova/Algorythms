Класс – конус. Методы – расчет площади и объема конуса. Поля – радиус основания, высота конуса.

```python
import math

class Cone():
    def __init__(self, R, h):
        self.R=R
        self.h=h
    
    def area_calculation(self):
        area = math.pi*self.R**2 + math.sqrt(self.h**2+self.R**2)*math.pi*self.R
        return area
    
    def volume_calculation(self):
        volume = (1/3)*((math.pi)*self.R**2)*self.h
        return volume

print('Введите радиус основания и высоту конуса:')
cone1=Cone(int(input()), int(input()))
print(f'Площадь конуса = {cone1.area_calculation()}, объем = {cone1.volume_calculation()}')
```

Опишите класс Triangle, заданный длиной одной стороны и двумя прилежащими углами (в градусах). Включите в описание класса методы, позволяющие вычислить две другие стороны и третий угол треугольника, и свойство, доступное только для чтения, позволяющее установить вид треугольника (равносторонний, равнобедренный, прямоугольный и т. п.).

```python
import math

class Triangle():
    def __init__(self, a, B, C):
        self.a=a
        self.B=B
        self.C=C
        self.A=180-(self.B+self.C)
        self.b=abs(self.a*math.sin(self.B)/math.sin(self.A))
        self.c=abs(self.a*math.sin(self.C)/math.sin(self.A))
        if self.a==self.b and self.b==self.c:
            self.__type='равносторонний'
        elif self.a==self.b or self.b==self.c or self.a==self.c:
            self.__type='равнобедренный'
        elif self.A==90 or self.B==90 or self.C==90:
            self.__type='прямоугольный'
        else:
            self.__type='разносторонний'
        
    def A_calculate(self):
        return self.A
        
    def b_calculate(self):
        return self.b        
        
    def c_calculate(self):
        return self.c

    def triangle_type(self):
        return self.__type
    

t1=Triangle(5, 90, 30)
t2=Triangle(6, 20, 20)
t3=Triangle(10, 60, 60)
t4=Triangle(11, 34, 46)
t_list=[t1, t2, t3, t4]
for i in t_list:
    print(f'Треугольник со стороной {i.a} и углами {i.B} и {i.C}')
    print('    Величина третьего угла: ', i.A_calculate())
    print('    Величина стороны b: ', i.b_calculate())
    print('    Величина стороны с: ', i.c_calculate())
    print('    Вид тругольника: ', i.triangle_type(), end='\n\n')
```

```python
def args_is_corr(func):
    def wrapper(*args):
        for arg in args:
            if not isinstance(arg, (int, float)):
                raise TypeError("Аргументом должно быть число типа int или float")
        return func(*args)
    return wrapper

@args_is_corr
def koren(a):
    return a**0.5

print('a)')
print(f'{koren(15):.3f}')
print(koren(15.21))
print(koren('artemiy'))
```

```python
def set_default_arguments(args_list, args_dict):
    def decorator(func):
        def wrapper(*args, **kwargs):
            all_args = list(args_list) + list(args_dict.values())
            for i, arg in enumerate(all_args):
                if i < len(args_list):
                    if i < len(args):
                        all_args[i] = args[i]
                    continue
                if i-len(args_list)<len(kwargs):
                    all_args[i]=list(kwargs.values())[i-len(args_list)]
            return func(*all_args)
        return wrapper
    return decorator

@set_default_arguments([1, 2], {'c': 3, 'd': 4})
def f(a, b, c=0, d=0):
    print(f"a={a}, b={b}, c={c}, d={d}")

print('б)')
f(11, 12)
f(12, c=34)
```

#### Дан стек. Необходимо удалить из него все отрицательные элементы.

Создать класс Stack с методами:

push(item): добавляет элемент item в верхушку стека;

pop(): удаляет и возвращает элемент из верхушки стека;

peek(): возвращает элемент из верхушки стека без его удаления;

is_empty(): проверяет, пуст ли стек;

size(): возвращает текущий размер стека/

Оба способа реализации должны обеспечивать функциональность стека: добавление элементов в верхушку, удаление элементов из верхушки, получение элемента из верхушки без его удаления, проверку на пустоту и получение текущего размера стека.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

def remove_negative_elements(stack):
    new_stack = Stack()
    while not stack.is_empty():
        item = stack.pop()
        if item >= 0:
            new_stack.push(item)
    return new_stack
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.size_count = 0

    def push(self, item):
        new_node = Node(item)
        new_node.next = self.top
        self.top = new_node
        self.size_count += 1

    def pop(self):
        if self.is_empty():
            return None
        popped_item = self.top.data
        self.top = self.top.next
        self.size_count -= 1
        return popped_item

    def peek(self):
        if self.is_empty():
            return None
        return self.top.data

    def is_empty(self):
        return self.top is None

    def size(self):
        return self.size_count

def remove_negative_elements(stack):
    new_stack = Stack()
    while not stack.is_empty():
        item = stack.pop()
        if item >= 0:
            new_stack.push(item)
    return new_stack
```

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        return self.items.pop(0)

    def peek(self):
        return self.items[0]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

def remove_duplicates(queue):
    unique_items = set()
    new_queue = Queue()
    while not queue.is_empty():
        item = queue.dequeue()
        if item not in unique_items:
            unique_items.add(item)
            new_queue.enqueue(item)
    return new_queue
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
        self.size_count = 0

    def enqueue(self, item):
        new_node = Node(item)
        if self.is_empty():
            self.front = new_node
        else:
            self.rear.next = new_node
        self.rear = new_node
        self.size_count += 1

    def dequeue(self):
        if self.is_empty():
            return None
        item = self.front.data
        self.front = self.front.next
        self.size_count -= 1
        return item

    def peek(self):
        if self.is_empty():
            return None
        return self.front.data

    def is_empty(self):
        return self.front is None

    def size(self):
        return self.size_count

def remove_duplicates(queue):
    unique_items = set()
    new_queue = Queue()
    while not queue.is_empty():
        item = queue.dequeue()
        if item not in unique_items:
            unique_items.add(item)
            new_queue.enqueue(item)
    return new_queue
```

Создайте двусвязный список для хранения информации о посетителях музея. Каждый элемент списка должен содержать имя и фамилию посетителя, дату посещения, список экспонатов, которые он посмотрел, и оценку каждого из них.

```python
class Node:
    def __init__(self, name, surname, visit_date):
        self.name = name
        self.surname = surname
        self.visit_date = visit_date
        self.exhibits = []
        self.next = None
        self.prev = None

class MuseumVisitorList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def add_visitor(self, name, surname, visit_date):
        new_node = Node(name, surname, visit_date)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
        self.size += 1

    def remove_visitor(self, name, surname):
        current = self.head
        while current:
            if current.name == name and current.surname == surname:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev
                self.size -= 1
                return
            current = current.next

    def get_visitor(self, name, surname):
        current = self.head
        while current:
            if current.name == name and current.surname == surname:
                return current
            current = current.next
        return None

    def get_exhibits(self, name, surname):
        visitor = self.get_visitor(name, surname)
        if visitor:
            return visitor.exhibits
        return []

    def add_exhibit(self, name, surname, exhibit):
        visitor = self.get_visitor(name, surname)
        if visitor:
            visitor.exhibits.append(exhibit)

    def get_rating(self, name, surname, exhibit):
        visitor = self.get_visitor(name, surname)
        if visitor:
            for exhibit in visitor.exhibits:
                if exhibit == exhibit:
                    return visitor.exhibits.index(exhibit)
        return None

    def set_rating(self, name, surname, exhibit, rating):
        visitor = self.get_visitor(name, surname)
        if visitor:
            for i, exhibit in enumerate(visitor.exhibits):
                if exhibit == exhibit:
                    visitor.exhibits[i] = (exhibit, rating)
                    return
        return

    def __str__(self):
        result = []
        current = self.head
        while current:
            result.append(f"{current.name} {current.surname} - {current.visit_date}")
            for exhibit, rating in current.exhibits:
                result.append(f"  - {exhibit}: {rating}")
            current = current.next
        return '\n'.join(result)
```

Реализовать функцию, которая находит сумму двух чисел, представленных в виде двусвязных списков.

```python
class Node:
    def __init__(self, data=0):
        self.data = data
        self.next = None
        self.prev = None

def sum_two_numbers(num1, num2):
    carry = 0
    result_head = Node()
    current = result_head

    while num1 or num2 or carry:
        val1 = num1.data if num1 else 0
        val2 = num2.data if num2 else 0

        total = val1 + val2 + carry
        carry = total // 10
        current.next = Node(total % 10)
        current.next.prev = current
        current = current.next

        if num1:
            num1 = num1.next
        if num2:
            num2 = num2.next

    result_head.next.prev = None
    return result_head.next

def create_linked_list_from_number(number):
    head = Node()
    current = head
    for digit in str(number)[::-1]:
        current.next = Node(int(digit))
        current.next.prev = current
        current = current.next
    head.next.prev = None
    return head.next

def print_linked_list(node):
    result = []
    while node:
        result.append(str(node.data))
        node = node.next
    print(''.join(result[::-1]))

num1 = create_linked_list_from_number(123)  # двусвязный список для числа 123
num2 = create_linked_list_from_number(456)  # двусвязный список для числа 456

result = sum_two_numbers(num1, num2)

print_linked_list(num1)
print_linked_list(num2)
print_linked_list(result)
```

Реализовать функцию, которая удаляет все элементы циклического двусвязного списка, большие заданного значения.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def remove_greater_elements(head, value):
    if not head or not head.next:
        return head

    last = head
    while last.next != head:
        last = last.next

    max_node = head
    current = head.next
    while current != head:
        if current.data > max_node.data:
            max_node = current
        current = current.next

    if max_node.data <= value:
        return head

    current = head
    while current.next != head:
        if current.next.data > value:
            next_node = current.next
            current.next = next_node.next
            next_node.prev = current
            if next_node == last:
                last = current
        else:
            current = current.next

    last.next = head
    head.prev = last

    return head
```

Необходимо отсортировать список слов по алфавиту и вывести результат на экран. В зависимости от переданного параметра отсортировать список слов по возрастанию или по убыванию алфавитного порядка, используя алгоритмы сортировки: сортировку вставками, сортировку выбором и быструю сортировку. Сравнить время выполнения алгоритмов сортировки с помощью декоратора. Текст хранится в файле.

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Время выполнения {func.__name__}: {end_time - start_time:.6f} секунд")
        return result
    return wrapper

def read_words_from_file(filename):
    with open(filename, 'r') as file:
        return file.read().splitlines()

@timer
def insertion_sort(words, reverse=False):
    for i in range(1, len(words)):
        key = words[i]
        j = i - 1
        while j >= 0 and (key.lower() < words[j].lower() if not reverse else key.lower() > words[j].lower()):
            words[j + 1] = words[j]
            j -= 1
        words[j + 1] = key
    return words

@timer
def selection_sort(words, reverse=False):
    for i in range(len(words)):
        min_idx = i
        for j in range(i + 1, len(words)):
            if (words[min_idx].lower() > words[j].lower() if not reverse else words[min_idx].lower() < words[j].lower()):
                min_idx = j
        words[i], words[min_idx] = words[min_idx], words[i]
    return words

def partition(words, low, high, reverse=False):
    pivot = words[(low + high) // 2].lower()
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while (words[i].lower() < pivot if not reverse else words[i].lower() > pivot):
            i += 1
        j -= 1
        while (words[j].lower() > pivot if not reverse else words[j].lower() < pivot):
            j -= 1
        if i >= j:
            return j
        words[i], words[j] = words[j], words[i]

@timer
def quick_sort(words, low=None, high=None, reverse=False):
    if low is None:
        low = 0
    if high is None:
        high = len(words) - 1
    if low < high:
        pivot = partition(words, low, high, reverse)
        quick_sort(words, low, pivot, reverse)
        quick_sort(words, pivot + 1, high, reverse)
    return words

if __name__ == "__main__":
    filename = "words.txt"
    words = read_words_from_file(filename)

    print("Сортировка по возрастанию:")
    insertion_sort(words.copy())
    selection_sort(words.copy())
    quick_sort(words.copy())

    print("\nСортировка по убыванию:")
    insertion_sort(words.copy(), reverse=True)
    selection_sort(words.copy(), reverse=True)
    quick_sort(words.copy(), reverse=True)
```

Реализовать класс бинарного дерева. Написать функцию для нахождения всех узлов, которые имеют только одного потомка в бинарном дереве.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(data, self.root)

    def _insert(self, data, node):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert(data, node.left)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert(data, node.right)

    def find_nodes_with_one_child(self):
        result = []
        self._find_nodes_with_one_child(self.root, result)
        return result

    def _find_nodes_with_one_child(self, node, result):
        if node:
            if (node.left and node.right is None) or (node.left is None and node.right):
                result.append(node.data)
            self._find_nodes_with_one_child(node.left, result)
            self._find_nodes_with_one_child(node.right, result)

tree = BinaryTree()
tree.insert(5)
tree.insert(3)
tree.insert(7)
tree.insert(2)
tree.insert(4)
tree.insert(6)
tree.insert(8)

nodes_with_one_child = tree.find_nodes_with_one_child()
print("Узлы с одним потомком:", nodes_with_one_child)
```

Имеется набор данных, представляющих собой пары «слово-частота». Реализуйте структуру данных на основе двоичной кучи, которая будет поддерживать операции вставки пары, удаления пары по слову и поиска частоты по слову.

```python
class WordFrequencyHeap:
    def __init__(self):
        self.heap = []
        self.word_index = {}

    def insert(self, word, frequency):
        self.heap.append((frequency, word))
        self.word_index[word] = len(self.heap) - 1
        self.heapify_up(len(self.heap) - 1)

    def delete(self, word):
        if word not in self.word_index:
            return

        index = self.word_index[word]
        self.swap(index, len(self.heap) - 1)
        self.heap.pop()
        del self.word_index[word]

        if index < len(self.heap):
            parent_index = (index - 1) // 2
            if self.heap[index][0] < self.heap[parent_index][0]:
                self.heapify_up(index)
            else:
                self.heapify_down(index)

    def search(self, word):
        if word not in self.word_index:
            return None
        index = self.word_index[word]
        return self.heap[index][0]

    def heapify_up(self, index):
        parent_index = (index - 1) // 2
        if index > 0 and self.heap[index][0] < self.heap[parent_index][0]:
            self.swap(index, parent_index)
            self.heapify_up(parent_index)

    def heapify_down(self, index):
        left_child_index = 2 * index + 1
        right_child_index = 2 * index + 2
        smallest = index

        if left_child_index < len(self.heap) and self.heap[left_child_index][0] < self.heap[smallest][0]:
            smallest = left_child_index

        if right_child_index < len(self.heap) and self.heap[right_child_index][0] < self.heap[smallest][0]:
            smallest = right_child_index

        if smallest != index:
            self.swap(index, smallest)
            self.heapify_down(smallest)

    def swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
        self.word_index[self.heap[i][1]] = i
        self.word_index[self.heap[j][1]] = j

heap = WordFrequencyHeap()
heap.insert("apple", 5)
heap.insert("banana", 3)
heap.insert("cherry", 7)
heap.insert("date", 2)

print(heap.search("banana"))
heap.delete("banana")
print(heap.search("banana"))
```

```python
# а)
class Country:
    def __init__(self, name, capital, population, area):
        self.name = name
        self.capital = capital
        self.population = population
        self.area = area
        
    def info(self):
        return print(f'{self.name}, {self.capital}, {self.population}, {self.area}')

class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.buckets = []

    def set(self, key, value):
        self.buckets.append([key, value])

    def get(self, key):
        index = self.hash(key)
        for pair in self.buckets[index]:
            if pair[0] == key:
                return pair[1]
        raise KeyError(key)

    def delete(self, key):
        index = self.hash(key)
        for i, pair in enumerate(self.buckets[index]):
            if pair[0] == key:
                del self.buckets[index][i]
                return
        raise KeyError(key)

country_table = HashTable()
country_table.set("Russia", Country("Russia", "Moscow", 144_100_000, 17_098_246))
country_table.set("USA", Country("USA", "Washington, D.C.", 328_239_523, 9_833_517))
country_table.set("China", Country("China", "Beijing", 1_439_323_776, 9_596_961))
for i in country_table.buckets:
    print(f'{i[0]}:')
    i[1].info()
```
